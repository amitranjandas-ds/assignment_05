# Normalization

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html
```

```python
from sklearn.preprocessing import MinMaxScaler
m = MinMaxScaler() #default feature_range=(0,1)
```

```python
df_ms = m.fit(df[['Marketing Spend']]) # calculation everything
```

```python
df_ms
```

```python
df.head()
```

```python
df[['Marketing_Spend_New']] = m.transform(df[['Marketing Spend']])
```

```python
df.head()
```

```python
df[['Administration']] = m.fit_transform(df[['Administration']])
df.head()
```

# Perform Normalization using loop

```python
df_n.head()
```

```python
for (columnName, columnData) in df_n.iteritems():    
    if(df_n[columnName].dtype == 'float64' and columnName != 'Profit'):
        df_n[[columnName]] = m.fit_transform(df_n[[columnName]])
df_n.head()
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
stds = StandardScaler()
```

```python
df_s.head()
```

```python
for (columnName, columnData) in df_s.iteritems():    
    if(df_s[columnName].dtype == 'float64' and columnName != 'Profit'):
        df_s[[columnName]] = stds.fit_transform(df_s[[columnName]])
df_s.head()
```

# Log Transformation

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.FunctionTransformer.html
```

```python
from sklearn.preprocessing import FunctionTransformer
ft = FunctionTransformer(np.log1p) #np.log1p will avoid 0
```

```python
df2.head(2)
```

```python
df2['Transport'] = ft.fit_transform(df2[['Transport']])
```

```python
df2.head()
```

# Perform Log using loop

```python
df_l.head()
```

```python
for (columnName, columnData) in df_l.iteritems():    
    if(df_l[columnName].dtype == 'float64' and columnName != 'Profit'):
        df_l[[columnName]] = ft.fit_transform(df_l[[columnName]])
df_l.head()
```
